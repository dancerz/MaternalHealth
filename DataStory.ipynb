1,539 - this is the number of counties in the US with zero obstetricians/gynecologists available per 100,000. For women in 1,539 counties, having a safe pregnancy becomes a luxury. Maternity healthcare stands as a fundamental pillar of a society's commitment to the well-being of its mothers and children. As defined by [March of Dimes](https://www.marchofdimes.org/maternity-care-deserts-report), Maternal Health Care deserts are areas that lack essential birthing clinics, OB/GYN, and midwife care within an accessible radius, resulting in infant and maternal mortality. With essentially ½ of US counties being classified as maternal healthcare deserts, it becomes a pressing issue worth investigating.

## Finding the Data
Our main data source for our research was the [U.S. Health Resources and Services Administration (HRSA), Area Health Resources Files (AHRF)](https://data.hrsa.gov/topics/health-workforce/ahrf). The U.S. Health Resources and Services Administration (HRSA) serves as a reputable and reliable source of healthcare-related data in the United States. The Area Health Resources Files (AHRF), a product of HRSA, collates data useful for assessing healthcare accessibility and resources across different regions. The credibility of HRSA stems comes from its status as a federal agency overseeing healthcare services and resources, ensuring the accuracy, reliability, and comprehensiveness of the data it provides. AHRF, specifically, collects, maintains, and updates diverse datasets, including information on healthcare professionals, facilities, health status indicators, and demographics. Researchers, policymakers, and healthcare professionals widely acknowledge and utilize HRSA's data to analyze healthcare disparities, plan interventions, and formulate evidence-based policies, underscoring the trustworthiness and credibility of this source.

We sourced the politcal data from the [MIT Election Data and Science Lab](https://electionlab.mit.edu/), more specifically, the "County Presidential Election Returns 2000-2020". We cleaned this data to focus on the 2020 election. The MIT Election Data and Science Lab has a reputation for providing comprehensive and reliable political data, serving as a reputable source for election-related information. The lab gathers and maintains a wide array of electoral data, including voting patterns, political affiliations, and election outcomes.

## The Politics of Maternal Healthcare

Maternal healthcare access is not only a matter of healthcare infrastructure but is deeply intertwined with political ideologies and governance. Understanding the role of federal and state governments in addressing this issue is crucial in dissecting the complexities behind maternal care deserts.

### Federal Government's Role
The federal government plays a pivotal role in setting national healthcare policies and frameworks, which significantly influence maternal health initiatives across the country. Recent federal initiatives, such as [The White House Blueprint for Addressing the Maternal Health Crisis](https://www.whitehouse.gov/wp-content/uploads/2022/06/Maternal-Health-Blueprint.pdf) which will be elaborated on later, illustrate the government's awareness of the challenges faced in maternal healthcare access.

### Party Perspectives
Within the political landscape, different ideologies shape approaches to healthcare policies, including maternity care. As explained by the [Pew Research Center](https://www.pewresearch.org/short-reads/2019/07/26/democrats-differ-over-best-way-to-provide-health-coverage-for-all-americans/), Democrats often prioritize expansive healthcare coverage, advocating for initiatives that increase accessibility, affordability, and quality of care. They endorse federal interventions and funding directed towards maternal health programs to bridge existing gaps and ensure equitable care for all expectant mothers. In contrast, Republicans' perspectives on healthcare tend to emphasize reduced federal intervention, favoring more localized solutions and state-level control over healthcare policies. This decentralized approach often intersects with varying levels of investment in maternal health infrastructure, resulting in disparities in care availability, particularly in regions leaning towards conservative ideologies.

This difference in ideologies is reflected in our data. Our analysis of county level data revealed that Texas, Missouri, Kansas, Nebraska, Kentucky, Georgia, Iowa, Virginia, Illinois, Oklahoma are the 10 states with the most number of counties having 0 Ob/Gyns per 100,000, in other words the most maternal healthcare deserts. At first glance, this may just appear to be an unfortunate chance. However, combining Ob/Gyn datasets with 2020 election data, we uncovered an intriguing correlation between political affiliations and the lack of maternity healthcare. The 10 states listed above are majority Republican states. This goes to suggest that perhaps the decentralized approach leaves some counties behind. Furthermore, it is important to note that Republican-leaning states, often championing "pro-life" rhetoric, paradoxically fall behind in providing the healthcare infrastructure necessary to support mothers through their pregnancy journeys. This reveals the contradiction between rhetoric and reality, and the double standards of being “pro-life”. In order for the lives of mothers and infants to truly be protected, maternity care deserts need to be addressed.




The pie charts above show the distribution of votes for counties considered maternal care deserts in the respective states. All counties that are maternal healthcare deserts in Missouri, Kansas, and Nebraska voted Republican in the 2020 election. In Georgia, 84% of the maternal healthcare desert counties voted Republican, which is still a majority considering Democrats won overall in the swing state of Georgia

### State vs. Federal Power
While federal policies and initiatives provide a broader framework, the power to implement and enforce changes in maternal healthcare predominantly resides within individual states. State governments have substantial autonomy in designing and executing healthcare programs and funding allocations. This decentralized approach results in a varied landscape of maternity care services, depending on each state's priorities, resources, and political inclinations.

In essence, the politics surrounding maternal healthcare access reflect a struggle between federal oversight and state autonomy, exacperated by ideological differences between political parties. Understanding these dynamics is essential in formulating comprehensive strategies to address maternal care deserts and bridge existing disparities in access to essential maternity services.

## Advocacy and the Future of Maternal Health

### Government Awareness



As briefly highlighted earlier, the June 2022 release of [The White House Blueprint for Addressing the Maternal Health Crisis](https://www.whitehouse.gov/wp-content/uploads/2022/06/Maternal-Health-Blueprint.pdf) suggests that the Biden Administration is aware of the situation. Starting with a letter from Vice President Kamala Harris, the pdf outlines 5 goals the administration hopes to achieve with regards to improving maternity care across the country. In line with our data analysis findings, the Blueprint also acknowledges that such circumstances disproportionately affect marginalized demographics such as Hispanic, indigenous, and Black mothers (pg 20). It argues that ensuring that everyone can reach and afford comprehensive maternal healthcare services is the first step to ending the crisis. The Blueprint outlines a comprehensive strategy to improve maternal healthcare, focusing on increasing access to high-quality services, particularly behavioral health services. Behavioral health, including perinatal depression, has been a significant concern in the US, which the COVID-19 pandemic has further exacerbated (pg 19). Untreated behavioral health conditions can contribute to pregnancy-associated deaths, underscoring the need for increased access to maternal healthcare during the prenatal and postpartum periods. The key actions identified to tackle these challenges include extending postpartum coverage, improving rural obstetric readiness, and expanding access to family planning services. Initiatives also focus on reducing uncontrolled hypertension, improving the quality of care for hypertensive disorders, and integrating behavioral health support in community settings (pg 21-26). The proposed measures aim to bridge existing gaps in maternal healthcare and promote a comprehensive and equitable approach to maternal well-being.

### Nebraska: A Maternal Healthcare Desert Making Small Strides
While in theory, federal government initiatives like the Blueprint provide a national framework, it is important to note that individual states are responsible for implementing this framework, and taking concrete actions to address the issue. In our Analysis, Nebraska was one of the states with the highest number of counties categorized as a maternal healthcare desert (0 ob/gyns per 100,000). Researching beyond our data, according to a March of Dime report entitled [2023 March of Dimes Report Card for Nebraska](https://www.marchofdimes.org/peristats/reports/nebraska/report-card), the state recognizes challenges in maternity care access and has taken steps to address them. An article from The North Platte Telegraph entitled [Nebraska Medicaid to be extended for moms up to 12 months after birth](https://nptelegraph.com/news/state-regional/government-politics/nebraska-medicaid-to-be-extended-for-moms-up-to-12-months-after-birth/article_effb7f0d-8d6d-5e5f-92cb-603736951526.html), Stoddard explains the pivotal nature of The Medicaid Presumptive Eligibility program, as it offers free healthcare coverage to eligible pregnant women while they await Medicaid application processing. Also highlighted in the article, in September 2023, Governor Jim Pillen announced to extend Nebraska Medicaid postpartum coverage from 60 days to 12 months, starting January 1, 2024. Providing full pregnancy-related Medicaid benefits during pregnancy and the extended postpartum period was mentioned as one of the key initiatives in the White House Blueprint, demonstrating the state's commitment and progress to comprehensive perinatal healthcare. The University of Nebraska Medical Center's Rural Workforce Preparation program is contributing to a sustainable solution as well. By providing education and preparation for the rural health workforce, the program is strategically designed to attract, educate, and graduate healthcare professionals committed to returning to Nebraska's rural areas, addressing the unique healthcare needs of these communities. This directly addresses the issue of a lack of Ob/Gyns highlighted by our data analysis.
